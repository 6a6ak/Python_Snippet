```python
import numpy as np
import matplotlib.pyplot as plt
data={
    'a': np.arange(50),
    'b': np.random.randint(0,50,50),
    'c':np.random.randn(50),
    'd': np.random.randint(500, size=50)  #random intiger number from 0-500 in array with size 50 
}
plt.scatter (data['a'],data['d'], color='#febd69')
plt.scatter (data['d'],data['b'], color='red')
plt.scatter (data['c'],data['d'], color='limegreen')
plt.ylabel('Y');plt.xlabel('X')
plt.show()
```
